```python
import numpy as np
import src.simulations as sim
import src.graph_utils as gutils
import src.utils as utils
import src.experiment_functions as expfuncs
import matplotlib.pyplot as plt
```

# n=500 pi=0.01

```python
pi = 0.01
initial_infection = [1]
# pc = 0.0001
# initial_cured = [400]
n = 500
m = 1000
spread_type = "mi-ss"
```

```python
print("Erdos-Renyi")
binomial = []
for i in range(10):
    gutils.generate_nx_graph("erdos-renyi", "graph.csv", n=500, m=1000)
    g = gutils.load_graph("graph.csv")
    res = expfuncs.run_simulation_over_graph(g, pi, number_of_runs=20, n_initial_infection=2, spread_type=spread_type)
    binomial.append(res)
    print(f"Graph {i} is completed")    
```

```python
print("Scale-free")
scalefree = []
for i in range(10):
    gutils.generate_nx_graph("scale-free", "graph.csv", n=500, m=1000)
    g = gutils.load_graph("graph.csv")
    res = expfuncs.run_simulation_over_graph(g, pi, number_of_runs=20, n_initial_infection=2, spread_type=spread_type)
    scalefree.append(res)
    print(f"Graph {i} is completed")        
```

```python
print("Small-worlds")
smallworlds03 = []
for i in range(10):
    gutils.generate_nx_graph("small-worlds", "graph.csv", n=500, m=1000, rewire_probability=0.3)
    g = gutils.load_graph("graph.csv")
    res = expfuncs.run_simulation_over_graph(g, pi, number_of_runs=20, n_initial_infection=2)
    smallworlds03.append(res)
    print(f"Graph {i} is completed")        
```

```python
def padded_stack(list_to_stack):
    max_size = max([len(a) for a in list_to_stack])
    padded_list = [np.pad(np.array(a),pad_width=(0,max_size-len(a)), mode='edge') for a in list_to_stack]
    return np.stack(padded_list)
```

```python
results = [binomial, scalefree,smallworlds03]#, smallworlds, smallworlds01]
labels = ["Erdos-Renyi", "Barabassi-Albert", "Watts-Strogatz p=0.3"]#, "Watts-Strogatz p=0.3",  "Watts-Strogatz p=0.1"]
for result in results:
    res_arr = [padded_stack(res[1]) for res in result]
    means = [np.mean(res, axis=0) for res in res_arr]
    mean = np.median(padded_stack(means), axis=0)
    std = np.std(padded_stack(means), axis=0)
    plt.plot(mean)    
    plt.fill_between(range(mean.shape[0]), mean-std, mean+std, alpha=0.3)  
    plt.legend(labels)
    
plt.xlabel("Simulation Step", size=12)
plt.ylabel("Number of Infected Nodes", size=12)
# plt.xlim([0,800])
plt.savefig("img/n500-p001-total.pdf", format='pdf')
```
